```python
# Import, Split, dan Normalisasi Data. Boleh diubah splitnya jika mau.
import tensorflow as tf
from tensorflow import keras
import numpy as np

mnist = tf.keras.datasets.mnist

(train_data, train_label), (test_data, test_label) = mnist.load_data()

train_data = train_data/255
test_data = test_data/255
```

```python
# Silakan menentukan arsitektur model sendiri

train_data = np.expand_dims(train_data, -1)
test_data = np.expand_dims(test_data, -1)
print("train_data shape:", train_data.shape)
print("test_data shape:", test_data.shape)
print(train_data.shape[0], "train_data samples")
print(test_data.shape[0], "test_data samples")

num_classes = 10
input_shape = (28, 28, 1)

train_label = tf.keras.utils.to_categorical(train_label, num_classes)
test_label = tf.keras.utils.to_categorical(test_label, num_classes)

model = tf.keras.Sequential(
    [
        tf.keras.Input(shape=input_shape),
        tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(num_classes, activation="softmax"),
    ]
)
model.summary()
```

```python
# Silakan menentukan loss dan optimizer sendiri
batch_size = 128
epochs = 15

model.compile(
    loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"]
)
```

```python
# Silakan menentukan banyaknya epoch sendiri

# Biar cantik aja progress bar nya
import sys
!{sys.executable} -m pip install tensorflow-addons
import tensorflow_addons as tfa
tqdm_callback = tfa.callbacks.TQDMProgressBar()

model.fit(
    train_data, train_label, batch_size=batch_size, epochs=epochs, validation_split=0.1,verbose=0, callbacks=[tqdm_callback]
)
```

```python
# Silakan evaluate test data
model.evaluate(
    test_data, test_label, verbose=0, callbacks=[tqdm_callback]
)
```
